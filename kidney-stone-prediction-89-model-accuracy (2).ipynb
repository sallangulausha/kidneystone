1) As per heatmap the feature urea and osmo strognly co-related

2) Feature osmo and cond also strongly co-related

3) Feature osmo and gravity also strongly co-related

4) Means we have problem of Multicollinearity hence, Logistic Regression may affect the performance of the model

```python
X = df.iloc[:,0:6].values
y = df['target'].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20, random_state=4)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
mms = MinMaxScaler()
X_train_scaled = mms.fit_transform(X_train)
X_test_scaled = mms.fit_transform(X_test)
```

# 4. Model Building(Logistic Regression, SVM, RandomForest,Lasso)

```python
# Let's implement Logistic Regression
model1 = LogisticRegression()
model1.fit(X_train_scaled,y_train)
y_pred1 = model1.predict(X_test_scaled)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
print(confusion_matrix(y_test,y_pred1))
```

```python
classes = ['AGGRESSIVE','NORMAL']
disp = ConfusionMatrixDisplay.from_predictions(y_test,y_pred1,display_labels=classes)
plt.rcParams['figure.figsize'] = [4,4]
plt.show()
```

```python
# Try out with SVM Model
from sklearn.svm import SVC
model2 = SVC(kernel = 'poly')
model2.fit(X_train_scaled,y_train)
y_pred2 = model2.predict(X_test_scaled)
```

```python
print(classification_report(y_test,y_pred2))
```

```python
print(confusion_matrix(y_test,y_pred2))
```

```python
# Let's check model performance with Ensemble techniques
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier()
model3.fit(X_train_scaled,y_train)
y_pred3 = model3.predict(X_test_scaled)
```

```python
print(classification_report(y_test,y_pred3))
```

```python
print(confusion_matrix(y_test,y_pred3))
```

```python
# as features are multicollinearity hence we have to Normalised featires by adding peanauty to coefficient 
from sklearn.linear_model import Lasso
model4 = Lasso(alpha = [0.005])
model4.fit(X_train_scaled,y_train)
y_pred4 = model4.predict(X_test_scaled)
```

 1) Here Alpha parameter are taken by trail and error method i.e. [1,0.1,0.001,0.001]
 
 2) We can also Used LassoCV means Lasso Cross Validation to obtained the best hyperparameter for this model 


```python
y_pred4
```

```python
# We have to Convert float to Binary class to compare between actual and predicted values
y_pred4_binary = np.round(y_pred4)
```

```python
y_pred4_binary
```

```python
print(classification_report(y_test,y_pred4_binary))
```

```python
print(confusion_matrix(y_test,y_pred4_binary))
```

```python
classes = ['AGGRESSIVE','NORMAL']
disp = ConfusionMatrixDisplay.from_predictions(y_test,y_pred4_binary,display_labels=classes)
plt.rcParams['figure.figsize'] = [4,4]
plt.show()
```

   Here,
   1) Our main goal is to check model accuracy rather than false positive and false negetive because both are equally important 

# 5. Inference

1) Logistic Regression perform well and 83% accuracy achieved 

2) SVM accuracy less due to high co-related features 

3) Random forest is less senstive to Multicollinierity but it is recommended to use regularisation techniques before implementing the model

4) Hence, We use L1 Regularization (Lasso Regression) to avoid multicollinearity by doing hyperparameter tunnning and achieved 89% accuracy of the model
